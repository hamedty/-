```python
import datetime
from urllib.request import urlretrieve
import requests


channel = 'tv1'
date = datetime.datetime.today().strftime('%Y-%m-%d')
hour = '1700'

m3u8_url = f'https://sdma.telewebion.com/vod/_definst_/media_b2/telewebion/{channel}/{date}/{hour}/file.mp4/playlist.m3u8?origin=pl2.telewebion.com'

m3u8 = requests.get(m3u8_url)
assert m3u8.status_code == 200, 'Telewebion m3u8 file denied'
m3u8 = [line for line in m3u8.text.splitlines() if line.startswith('http')]

for index, url in enumerate(m3u8):
    urlretrieve(url, f'videos/{index:03d}.ts')

#   cat videos/*.ts > videos/merged.ts
#   ffmpeg -i videos/merged.ts -ar 16000 -ac 1 -y videos/merged.wav

```

```python
from vosk import Model, KaldiRecognizer
import json

model = Model("model")
rec = KaldiRecognizer(model, 16000)

```

```python
wf = open('videos/merged.wav', "rb")
wf.read(44) # skip header

i = 0
while True:
    data = wf.read(320000)
    if len(data) == 0:
        break
    if rec.AcceptWaveform(data):
        res = json.loads(rec.Result())
        print (i, res['text'])
    i += 1
res = json.loads(rec.FinalResult())
print (res['text'])
print('------')
```
